```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from importlib import reload
import warnings
from typing import Tuple
from continuous_probablistic_methods import get_lower_and_upper_bounds
warnings.simplefilter("ignore")
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
lemonade = pd.read_csv('lemonade.csv')
cont_lemonade = lemonade[['Temperature','Rainfall','Flyers','Sales']]
```

```python
lemonade.head()
```

```python
bounds = {}
for s in ['Temperature','Rainfall','Flyers','Sales']:
    series = lemonade[s]
    bounds[s] = get_lower_and_upper_bounds(series)
bounds
```

```python
dfs = []
for name,bds in bounds.items():
    dfs.append(lemonade[(lemonade[name] < bds[0]) | (lemonade[name] > bds[1])])
pd.concat(dfs)
```
